```python
def read_documents(directory):
    documents = {}
    doc_files = sorted([f for f in os.listdir(directory) if f.startswith('D') and f.endswith('.txt')])
    
    for doc_file in doc_files:
        file_path = os.path.join(directory, doc_file)
        with open(file_path, 'r') as f:
            documents[doc_file] = f.read().strip()
    
    return documents

def generate_char_2grams(text):
    grams = set()
    for i in range(len(text) - 1):
        grams.add(text[i:i+2])
    return grams

def generate_char_3grams(text):
    grams = set()
    for i in range(len(text) - 2):
        grams.add(text[i:i+3])
    return grams

def generate_word_2grams(text):
    words = text.split()
    grams = set()
    for i in range(len(words) - 1):
        grams.add((words[i], words[i+1]))
    return grams
```

```python
import hashlib

def generate_3grams(text):
    grams = set()
    for i in range(len(text) - 2):
        grams.add(text[i:i+3])
    return grams

def hash_function(gram, seed):
    hash_obj = hashlib.md5((str(seed) + gram).encode())
    return int(hash_obj.hexdigest(), 16)

def compute_min_hash_signature(grams, num_hash_functions):
    signature = []
    for seed in range(num_hash_functions):
        min_hash = float('inf')
        for gram in grams:
            h = hash_function(gram, seed)
            if h < min_hash:
                min_hash = h
        signature.append(min_hash)
    return signature

def estimate_jaccard_similarity(sig1, sig2):
    if len(sig1) == 0 or len(sig2) == 0:
        return 0.0
    matches = sum(1 for h1, h2 in zip(sig1, sig2) if h1 == h2)
    return matches / len(sig1)

d1_text = documents['D1.txt']
d2_text = documents['D2.txt']

d1_3grams = generate_3grams(d1_text)
d2_3grams = generate_3grams(d2_text)

print("="*80)
print("MIN-HASH SIGNATURE ANALYSIS FOR D1 AND D2")
print("="*80)
print(f"\nD1.txt - Unique 3-grams: {len(d1_3grams)}")
print(f"D2.txt - Unique 3-grams: {len(d2_3grams)}")
print(f"Intersection: {len(d1_3grams & d2_3grams)}")
print(f"Union: {len(d1_3grams | d2_3grams)}")

true_jaccard = len(d1_3grams & d2_3grams) / len(d1_3grams | d2_3grams) if len(d1_3grams | d2_3grams) > 0 else 0
print(f"\nTrue Jaccard Similarity: {true_jaccard:.4f}")

print("\n" + "="*80)
print("APPROXIMATE JACCARD SIMILARITY USING MIN-HASH")
print("="*80)

t_values = [20, 60, 150, 300, 600]
approximate_similarities = []

for t in t_values:
    sig_d1 = compute_min_hash_signature(d1_3grams, t)
    sig_d2 = compute_min_hash_signature(d2_3grams, t)
    
    approx_jaccard = estimate_jaccard_similarity(sig_d1, sig_d2)
    approximate_similarities.append(approx_jaccard)
    
    print(f"\nt = {t:3d} hash functions:")
    print(f"  Approximate Jaccard Similarity: {approx_jaccard:.4f}")
    print(f"  Error from true Jaccard: {abs(approx_jaccard - true_jaccard):.4f}")

print("\n" + "="*80)
print("SUMMARY - APPROXIMATE JACCARD SIMILARITIES")
print("="*80)
for t, sim in zip(t_values, approximate_similarities):
    print(f"t = {t:3d}: {sim:.4f}")
```

```python
import hashlib

m = 100000

def hash_to_range(gram, seed, m):
    hash_obj = hashlib.md5((str(seed) + gram).encode())
    hash_value = int(hash_obj.hexdigest(), 16)
    return hash_value % m

def compute_min_hash_signature_part_a(grams, num_hash_functions, m):
    signature = []
    for seed in range(num_hash_functions):
        min_hash = float('inf')
        for gram in grams:
            h = hash_to_range(gram, seed, m)
            if h < min_hash:
                min_hash = h
        signature.append(min_hash)
    return signature

def estimate_jaccard_from_signature(sig1, sig2):
    if len(sig1) == 0 or len(sig2) == 0:
        return 0.0
    matches = sum(1 for h1, h2 in zip(sig1, sig2) if h1 == h2)
    return matches / len(sig1)

d1_text = documents['D1.txt']
d2_text = documents['D2.txt']

d1_3grams = generate_3grams(d1_text)
d2_3grams = generate_3grams(d2_text)

print("="*80)
print("PART 2A: MIN-HASH SIGNATURES FOR D1 AND D2 (3-GRAMS)")
print("="*80)
print(f"\nHash Family: h : k-grams → [m] where m = {m}")
print(f"\nD1.txt - Unique 3-grams: {len(d1_3grams)}")
print(f"D2.txt - Unique 3-grams: {len(d2_3grams)}")
print(f"Intersection: {len(d1_3grams & d2_3grams)}")
print(f"Union: {len(d1_3grams | d2_3grams)}")

true_jaccard = len(d1_3grams & d2_3grams) / len(d1_3grams | d2_3grams)
print(f"\nTrue Jaccard Similarity: {true_jaccard:.4f}")

print("\n" + "="*80)
print("APPROXIMATE JACCARD SIMILARITY USING MIN-HASH")
print("="*80)

t_values = [20, 60, 150, 300, 600]
approximate_jaccard_values = []

for t in t_values:
    sig_d1 = compute_min_hash_signature_part_a(d1_3grams, t, m)
    sig_d2 = compute_min_hash_signature_part_a(d2_3grams, t, m)
    
    approx_jaccard = estimate_jaccard_from_signature(sig_d1, sig_d2)
    approximate_jaccard_values.append(approx_jaccard)
    
    print(f"\nt = {t:3d} hash functions:")
    print(f"  Approximate Jaccard Similarity: {approx_jaccard:.4f}")
    print(f"  Error from true value: {abs(approx_jaccard - true_jaccard):.4f}")

print("\n" + "="*80)
print("FINAL ANSWER - PART 2A: 5 APPROXIMATE JACCARD SIMILARITIES")
print("="*80)
print("\nFor document pair (D1, D2) using 3-grams:")
print()
for t, sim in zip(t_values, approximate_jaccard_values):
    print(f"  t = {t:3d}:  {sim:.4f}")
```

```python
import numpy as np

t = 160
tau = 0.7

def s_curve(s, b, r):
    """
    Compute the LSH S-curve probability: f(s) = 1 - (1 - s^b)^r
    This gives the probability that two items with Jaccard similarity s
    will hash to the same bucket in at least one band.
    """
    if s == 0:
        return 0.0
    if s == 1:
        return 1.0
    return 1.0 - (1.0 - (s ** b)) ** r

def find_divisors(n):
    """Find all divisor pairs (b, r) where b * r = n"""
    divisors = []
    for i in range(1, int(np.sqrt(n)) + 1):
        if n % i == 0:
            b = i
            r = n // i
            divisors.append((b, r))
            if i != n // i:
                b = n // i
                r = i
                divisors.append((b, r))
    return sorted(divisors)

def evaluate_s_curve_quality(b, r, tau):
    """
    Evaluate the quality of the S-curve for LSH.
    We want:
    - Sharp transition around tau
    - f(tau) close to 0.5 (ideally)
    - High probability for similar pairs (s >> tau)
    - Low probability for dissimilar pairs (s << tau)
    """
    f_tau = s_curve(tau, b, r)
    f_high = s_curve(tau + 0.2, b, r)  
    f_low = s_curve(tau - 0.2, b, r)   
    
    separation = f_high - f_low
    distance_from_half = abs(f_tau - 0.5)
    
    return {
        'f_tau': f_tau,
        'f_high': f_high,
        'f_low': f_low,
        'separation': separation,
        'distance_from_half': distance_from_half
    }

print("="*80)
print("PART 3A: LSH BAND CONFIGURATION ANALYSIS")
print("="*80)
print(f"\nParameters:")
print(f"  Total hash functions (t): {t}")
print(f"  Similarity threshold (τ): {tau}")
print(f"  Formula: f(s) = 1 - (1 - s^b)^r")
print(f"  Constraint: b × r = {t}")

divisors = find_divisors(t)
print(f"\nFound {len(divisors)} possible (b, r) configurations\n")

results_lsh = []
for b, r in divisors:
    quality = evaluate_s_curve_quality(b, r, tau)
    results_lsh.append({
        'b': b,
        'r': r,
        **quality
    })

print("="*80)
print("LSH S-CURVE ANALYSIS FOR ALL (b, r) CONFIGURATIONS")
print("="*80)
print(f"\n{'b':>4} | {'r':>4} | {'f(0.5)':>7} | {'f(0.7)':>7} | {'f(0.9)':>7} | {'Separation':>11} | {'Quality':>7}")
print("-" * 80)

for res in sorted(results_lsh, key=lambda x: -x['separation']):
    f_low = s_curve(0.5, res['b'], res['r'])
    f_high = s_curve(0.9, res['b'], res['r'])
    quality_score = res['separation'] - res['distance_from_half'] * 0.5
    print(f"{res['b']:>4} | {res['r']:>4} | {f_low:>7.4f} | {res['f_tau']:>7.4f} | {f_high:>7.4f} | {res['separation']:>11.4f} | {quality_score:>7.4f}")

print("\n" + "="*80)
print("RECOMMENDED CONFIGURATIONS")
print("="*80)

top_configs = sorted(results_lsh, key=lambda x: -x['separation'])[:5]

for i, res in enumerate(top_configs, 1):
    print(f"\n{i}. b = {res['b']}, r = {res['r']}")
    print(f"   S-curve values:")
    print(f"      f(0.5) = {s_curve(0.5, res['b'], res['r']):.4f} (dissimilar pairs, want LOW)")
    print(f"      f(0.7) = {res['f_tau']:.4f} (threshold)")
    print(f"      f(0.9) = {s_curve(0.9, res['b'], res['r']):.4f} (similar pairs, want HIGH)")
    print(f"   Separation: {res['separation']:.4f}")
    print(f"   Interpretation: At τ={tau}, this configuration hashes similar pairs with")
    print(f"                   probability {res['f_tau']:.1%} and dissimilar pairs with")
    print(f"                   probability {res['f_low']:.1%}")

print("\n" + "="*80)
print("FINAL ANSWER - RECOMMENDED (b, r) VALUES")
print("="*80)

best = max(results_lsh, key=lambda x: x['separation'])
print(f"\n✓ PRIMARY RECOMMENDATION:")
print(f"  b = {best['b']:3d} (rows per band)")
print(f"  r = {best['r']:3d} (number of bands)")
print(f"  b × r = {best['b'] * best['r']} (total hash functions)")
print(f"\n  S-curve at threshold τ = {tau}:")
print(f"    f(τ) = {best['f_tau']:.4f}")
print(f"    Probability similar pairs collide: {best['f_high']:.1%}")
print(f"    Probability dissimilar pairs collide: {best['f_low']:.1%}")
print(f"    Separation quality: {best['separation']:.4f}")

print(f"\n✓ ALTERNATIVE RECOMMENDATIONS (for different tradeoffs):")
for res in top_configs[1:3]:
    print(f"  • b = {res['b']:3d}, r = {res['r']:3d}: Separation = {res['separation']:.4f}")
```

```python
print("\n" + "="*80)
print("STEP 1: COMPUTE EXACT JACCARD SIMILARITIES")
print("="*80)

def compute_jaccard_movies(user1_movies, user2_movies):
    """Compute Jaccard similarity between two users' movie sets"""
    if len(user1_movies) == 0 and len(user2_movies) == 0:
        return 1.0
    intersection = len(user1_movies & user2_movies)
    union = len(user1_movies | user2_movies)
    return intersection / union if union > 0 else 0.0

exact_similarities = {}
exact_pairs_above_threshold = []

print(f"\nComputing exact Jaccard similarities for {len(user_ids)} users...")
print(f"Total pairs to compute: {len(user_ids) * (len(user_ids) - 1) // 2}")

for i in range(len(user_ids)):
    for j in range(i + 1, len(user_ids)):
        user1 = user_ids[i]
        user2 = user_ids[j]
        
        sim = compute_jaccard_movies(user_movies[user1], user_movies[user2])
        exact_similarities[(user1, user2)] = sim
        
        if sim >= 0.5:
            exact_pairs_above_threshold.append((user1, user2, sim))

exact_pairs_above_threshold.sort(key=lambda x: -x[2])

print(f"Pairs with Jaccard similarity >= 0.5: {len(exact_pairs_above_threshold)}")

if len(exact_pairs_above_threshold) > 0:
    print(f"\nTop 10 similar user pairs:")
    for user1, user2, sim in exact_pairs_above_threshold[:10]:
        print(f"  Users {user1} <-> {user2}: {sim:.4f}")

exact_similar_set = set((u1, u2) for u1, u2, _ in exact_pairs_above_threshold)

print(f"\nTotal exact similar pairs (Jaccard >= 0.5): {len(exact_similar_set)}")
```

```python

import random
import math

def next_prime(n):
    def is_prime(x):
        if x < 2:
            return False
        if x % 2 == 0:
            return x == 2
        r = int(math.sqrt(x))
        for i in range(3, r + 1, 2):
            if x % i == 0:
                return False
        return True
    p = n + 1
    while not is_prime(p):
        p += 1
    return p


max_movie = max(m for movies in user_movies.values() for m in movies)
def make_hash_funcs(k, max_item, seed=None):
    if seed is not None:
        rnd = random.Random(seed)
    else:
        rnd = random
    prime = next_prime(max_item * 10 + 1)
    funcs = []
    for _ in range(k):
        a = rnd.randint(1, prime - 1)
        b = rnd.randint(0, prime - 1)
        funcs.append((a, b, prime))
    return funcs

def apply_hash(a, b, p, x):
    return (a * x + b) % p

def compute_signatures(users, user_sets, funcs):
    signatures = {u: [None] * len(funcs) for u in users}
    for idx, (a, b, p) in enumerate(funcs):
        for u in users:
            items = user_sets[u]
            minv = None
            for it in items:
                h = apply_hash(a, b, p, it)
                if minv is None or h < minv:
                    minv = h
            if minv is None:
                minv = p
            signatures[u][idx] = minv
    return signatures

def estimate_pairs_from_sigs(users, signatures, threshold=0.5):
    pairs = set()
    k = len(next(iter(signatures.values())))
    for i in range(len(users)):
        u = users[i]
        su = signatures[u]
        for j in range(i+1, len(users)):
            v = users[j]
            sv = signatures[v]
            eq = sum(1 for t in range(k) if su[t] == sv[t])
            if eq / k >= threshold:
                pairs.add((u, v))
    return pairs


print("\n" + "="*80)
print("STEP 2: MIN-HASH EXPERIMENTS")
print("="*80)

true_pairs = exact_similar_set
results = {}
for k in [50, 100, 200]:
    run_stats = []
    for run in range(5):
        funcs = make_hash_funcs(k, max_movie, seed=run)
        sigs = compute_signatures(user_ids, user_movies, funcs)
        est = estimate_pairs_from_sigs(user_ids, sigs, threshold=0.5)
        fp = len([p for p in est if p not in true_pairs])
        fn = len([p for p in true_pairs if p not in est])
        run_stats.append((len(est), fp, fn))
    avg_est = sum(r[0] for r in run_stats)/5
    avg_fp = sum(r[1] for r in run_stats)/5
    avg_fn = sum(r[2] for r in run_stats)/5
    results[k] = (run_stats, avg_est, avg_fp, avg_fn)
    print(f"k={k}: avg_est={avg_est:.2f}, avg_fp={avg_fp:.2f}, avg_fn={avg_fn:.2f}")

print("\nDone experiments.")
```

```python


def lsh_candidate_pairs(signatures, r, b):
    """Return set of user pairs that collide in at least one of the r bands.
    signatures: dict user -> list of hash values
    r: number of bands, b: rows per band (r * b == len(signature))
    """
    buckets = {}
    candidates = set()
    users = list(signatures.keys())
    for u in users:
        sig = signatures[u]
        for band in range(r):
            start = band * b
            end = start + b
            band_tuple = tuple(sig[start:end])
            key = (band, hash(band_tuple))
            buckets.setdefault(key, []).append(u)
    for bucket_users in buckets.values():
        if len(bucket_users) > 1:
            for i in range(len(bucket_users)):
                for j in range(i+1, len(bucket_users)):
                    a = bucket_users[i]
                    b_u = bucket_users[j]
                    if a < b_u:
                        candidates.add((a, b_u))
                    else:
                        candidates.add((b_u, a))
    return candidates


configurations = {
    50: [(5, 10)],
    100: [(5, 20)],
    200: [(5, 40), (10, 20)]
}

for thresh in [0.6, 0.8]:
    true_pairs_thresh = {pair for pair, s in exact_similarities.items() if s >= thresh}
    print("\n" + "="*80)
    print(f"THRESHOLD = {thresh}")
    print(f"True pairs (>= {thresh}): {len(true_pairs_thresh)}")
    print("="*80)
    
    for k, cfgs in configurations.items():
        print(f"\n-- k = {k} signature length --")
        for (r, b) in cfgs:
            assert r * b == k, "band-row product must equal signature size"
            stats = []
            for run in range(5):
                funcs = make_hash_funcs(k, max_movie, seed=run)
                sigs = compute_signatures(user_ids, user_movies, funcs)
                cand = lsh_candidate_pairs(sigs, r, b)
                fp = len([p for p in cand if p not in true_pairs_thresh])
                fn = len([p for p in true_pairs_thresh if p not in cand])
                stats.append((len(cand), fp, fn))
            avg_est = sum(x[0] for x in stats) / 5
            avg_fp = sum(x[1] for x in stats) / 5
            avg_fn = sum(x[2] for x in stats) / 5
            print(f"r={r}, b={b}: avg_candidates={avg_est:.1f}, avg_fp={avg_fp:.1f}, avg_fn={avg_fn:.1f}")
    
print("\nLSH experiments complete.")
```
